```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('life-expectancy-vs-gdp.csv')
data.rename({'Population (historical estimates)':'Population_in_Millions'},axis=1,inplace=True)
data['Population_in_Millions']=np.round(data['Population_in_Millions']/1000000,2)
data.head()
```

```python
data[data['Year']==2015]
```

```python
df=data[['Life expectancy','GDP per capita','Population_in_Millions','Continent']]
df=df.dropna()

df.head()
```

```python
df.isna().sum()
df.shape
```

```python
# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(df['Life expectancy'],df['GDP per capita'])
plt.show()
```

```python
plt.scatter(df['Life expectancy'],df['GDP per capita'])

plt.show()
```

```python
population=np.array(df['Population_in_Millions'])
pop=population*2
plt.scatter(df['Life expectancy'],df['GDP per capita'],s=pop)
plt.show()
```

```python
plt.hist(df['Life expectancy'])
plt.show()
```

#### Data might not be accurate. Code just for reference of plots

```python
#plt.scatter(df['GDP per capita'],df['Life expectancy'])
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)


# Add title
plt.title(title)

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val,tick_lab)

# Update: set s argument to np_pop
population=np.array(df['Population_in_Millions'])
pop=population*2


# Addding colors according to continents
col = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'South America':'yellow',
    'North America':'orange',
    'Oceania':'black'
}
df['color']=df['Continent'].map(col)
col=np.array(df['color'])

plt.scatter(df['Life expectancy'],df['GDP per capita'],s=pop,c=col)

plt.text(65,25000,'Asia')

# After customizing, display the plot
plt.show()
```
